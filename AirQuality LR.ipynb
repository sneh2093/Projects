## Air Quality Data Set
Data Set Information:
Source: https://archive.ics.uci.edu/ml/datasets/Air+quality#

The dataset contains 9358 instances of hourly averaged responses from an array of 5 metal oxide chemical sensors embedded in an Air Quality Chemical Multisensor Device. The device was located on the field in a significantly polluted area, at road level,within an Italian city. Data were recorded from March 2004 to February 2005 (one year)representing the longest freely available recordings of on field deployed air quality chemical sensor devices responses. Ground Truth hourly averaged concentrations for CO, Non Metanic Hydrocarbons, Benzene, Total Nitrogen Oxides (NOx) and Nitrogen Dioxide (NO2) and were provided by a co-located reference certified analyzer. Evidences of cross-sensitivities as well as both concept and sensor drifts are present as described in De Vito et al., Sens. And Act. B, Vol. 129,2,2008 (citation required) eventually affecting sensors concentration estimation capabilities. Missing values are tagged with -200 value.
This dataset can be used exclusively for research purposes. Commercial purposes are fully excluded.


Attribute Information:

0 &nbsp; Date (DD/MM/YYYY) <br>
1 &nbsp; Time (HH.MM.SS) <br>
2 &nbsp; True hourly averaged concentration CO in mg/m^3 (reference analyzer) <br>
3 &nbsp; PT08.S1 (tin oxide) hourly averaged sensor response (nominally CO targeted) <br>
4 &nbsp; True hourly averaged overall Non Metanic HydroCarbons concentration in microg/m^3 (reference analyzer) <br>
5 &nbsp; True hourly averaged Benzene concentration in microg/m^3 (reference analyzer) <br>
6 &nbsp; PT08.S2 (titania) hourly averaged sensor response (nominally NMHC targeted) <br>
7 &nbsp; True hourly averaged NOx concentration in ppb (reference analyzer) <br>
8 &nbsp; PT08.S3 (tungsten oxide) hourly averaged sensor response (nominally NOx targeted) <br>
9 &nbsp; True hourly averaged NO2 concentration in microg/m^3 (reference analyzer) <br>
10 &nbsp; PT08.S4 (tungsten oxide) hourly averaged sensor response (nominally NO2 targeted) <br>
11 &nbsp; PT08.S5 (indium oxide) hourly averaged sensor response (nominally O3 targeted) <br>
12 &nbsp; Temperature in Â°C <br>
13 &nbsp; Relative Humidity (%) <br>
14 &nbsp; AH Absolute Humidity <br>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### Importing and Analyzing the Data

```python
data=pd.read_csv('AirQualityUCI.csv',';')
print("The Dataset contain {} Rows and {} Columns".format(data.shape[1], data.shape[0]))
display(data.head(10))
display(data.info())
display(data.describe())
print("Count of Data Types:")
display(data.dtypes.value_counts())
```

#### Cleaning the Data by removing Unnamed Columns and Rounding of all float values.

```python
data.drop('Unnamed: 15', axis=1 , inplace=True)
data.drop('Unnamed: 16', axis=1 , inplace=True)
data=data.round(decimals=2)
data.head(10)
```

#### Seperating the Numeric and String Columns
Also checking to see if any of the numeric columns were missing data

```python
num_var=data.columns[data.dtypes!=object]
char_var=data.columns[data.dtypes==object]
data[num_var].isnull()
```

Dropping all columns with missing data.

```python
data=data.dropna()
data[num_var].isnull().value_counts()
```

#### Linear Regression

```python
X=data[[ 'NMHC(GT)','PT08.S2(NMHC)', 'NOx(GT)', 'PT08.S3(NOx)', 'NO2(GT)', 'PT08.S4(NO2)','PT08.S5(O3)']]
Y= data['PT08.S1(CO)']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test= train_test_split(X,Y,test_size=0.4, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
lm.fit(X_train, Y_train)
```

```python
print(lm.intercept_)
```

```python
lm.coef_
```

```python
cdf=pd.DataFrame(lm.coef_,X.columns,columns=['Coeff'])
cdf
```

```python
predictions= lm.predict(X_test)
```

```python
predictions
```

```python
plt.scatter(Y_test, predictions)
```

```python
sns.distplot(Y_test-predictions)
```
